```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
iris
```

```python
iris.data
```

```python
iris.target
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
# independent and dependent features
X=df.iloc[:,:-1]
Y=iris.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(
X,Y,test_size=0.33, random_state = 42)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn import tree
```

```python
# postpurning
tre = DecisionTreeClassifier(max_depth=2)
tre.fit(X_train, Y_train)
```

```python
plt.figure(figsize=(16,15))
#tree.plot_tree(tre,filled=True)
plot_tree(tre, filled=True, feature_names=X_train.columns, class_names=['Setosa', 'Versicolor', 'Virginica'])
plt.show()
```

```python
# Hyperparameter = max_depth = 2
```

```python
y_pred = tre.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score=accuracy_score(y_pred,Y_test)
print(score)
```

```python
print(classification_report(y_pred,Y_test))
```
